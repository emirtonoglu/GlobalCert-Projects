### Number of actions

```python
bidders =  bids['bidder_id'].unique()
bidders_bids_list = bids[['bidder_id','auction']].groupby('bidder_id').count()['auction']
def numberofActions(a,b,c):
    if not a['bidder_id'] in c:
        return 0
    else:
        return b[a['bidder_id']]
train['Number of Actions'] = train.apply(lambda x: numberofActions(x,bidders_bids_list,bidders),axis=1) 
train
```

### Time Response

```python
lastaction={}   #dictionary for last action

def timer(a,b):
    if a['auction'] in b:
        time = a['time'] - b[a['auction']]
        b[a['auction']] = a['time']
        return time
    else :
        b[a['auction']] = a['time']
        return 0

bids['timeresponse'] = bids.apply(lambda x: timer(x,lastaction),axis=1)
bids.head(5)
```

```python
grouped_timeresponse = bids[['bidder_id','timeresponse']].groupby('bidder_id').mean()['timeresponse']
def meantimeresponse(a,b,c):
    if not a['bidder_id'] in c:
        return 0
    else:
        return b[a['bidder_id']]
    
train['Mean Time Response'] = train.apply(lambda x: meantimeresponse(x,grouped_timeresponse,bidders),axis=1)
train
```

```python
import dataframe_image as dfi
temp = train.groupby(['outcome']).agg({'Mean Time Response': ['median','mean']})
dfi.export(temp, 'graph6.png')
```

```python
grouped_timeresponse = bids[['bidder_id','timeresponse']].groupby('bidder_id').min()['timeresponse']
def mintimeresponse(a,b,c):
    if not a['bidder_id'] in c:
        return 0
    else:
        return b[a['bidder_id']]
    
train['Min Time Response'] = train.apply(lambda x: mintimeresponse(x,grouped_timeresponse,bidders),axis=1)
train
```

```python
import dataframe_image as dfi
temp = train.groupby(['outcome']).agg({'Min Time Response': ['median','mean']})
dfi.export(temp, 'graph7.png')
```

### First and Last Bids

```python
last_bid = bids.sort_values(['auction', 'time'], ascending = [True, False])
last_bid = last_bid.groupby('auction').first().reset_index()
last_bid = last_bid.groupby('bidder_id').count()['bid_id'].reset_index()
last_bid = last_bid.rename(columns = {'bid_id': 'num_last_bid'})
first_bid = bids.sort_values(['auction', 'time'])
first_bid = first_bid.groupby('auction').first().reset_index()
first_bid = first_bid.groupby('bidder_id').count()['bid_id'].reset_index()
first_bid = first_bid.rename(columns = {'bid_id': 'num_first_bid'})
```

```python
train = train.merge(first_bid, on='bidder_id', how='left').fillna(0)
train = train.merge(last_bid, on='bidder_id', how='left').fillna(0)
```

```python
train = train.rename(columns={"num_first_bid" : "Number of First Bids"})
train = train.rename(columns={"num_last_bid" : "Number of Auctions Won"})
train['Number of First Bids'] = train['Number of First Bids'].astype(int)
train['Number of Auctions Won'] = train['Number of Auctions Won'].astype(int)
train
```

```python
import dataframe_image as dfi
temp = train.groupby(['outcome']).agg({'Number of First Bids': ['median','mean']})
dfi.export(temp, 'graph8.png')
```

### Model

```python
x = train.drop(['bidder_id','outcome'],axis=1)
y = train.outcome
```

```python
from sklearn.model_selection import train_test_split

x_train, x_cv, y_train, y_cv = train_test_split(x,y, test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier

from sklearn.model_selection import cross_val_score

rand = RandomForestClassifier(n_estimators=600,max_depth=15,min_samples_leaf=2)
print (cross_val_score(rand,x_train,y_train,cv=3).mean())
```

```python
rand.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
pred_cv = rand.predict(x_cv)
accuracy_score(y_cv,pred_cv)
```

```python
from sklearn.metrics import confusion_matrix
cnf=confusion_matrix(y_cv, pred_cv)
cnf
```

```python

classes = ["0", "1"]
df_cnf = pd.DataFrame(cnf, index = classes, columns = classes)
plt.figure(figsize = (10,7))
cnf_plot = sns.heatmap(df_cnf, cmap ="Greens" ,annot=True)
cnf_plot.figure.savefig("cnf.png")
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier(random_state=1)
model.fit(x_train,y_train)
```

```python
pred_cv = model.predict(x_cv)
accuracy_score(y_cv,pred_cv)
```
